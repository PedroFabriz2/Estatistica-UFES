```python
#Medidas Resumo dos dados v6

Intervalos  = [4.9, 5.1, 5.3, 5.5, 5.7, 5.9, 6.1, 6.3, 6.5, 6.7, 6.9]
representantes = np.zeros(len(Intervalos)-1)

#criacao dos numeros representantes para calculo das medidas resumo
for i in range(1, len(Intervalos), 1):
    representantes[i-1] = (Intervalos[i-1] + Intervalos[i])/2
    
df6_medidas = df6[["V6(faixa)"]].drop(10, 0)
```

```python
df6_medidas["Representantes"] = representantes
```

```python
representantes
```

```python
df6_medidas["Frequencia"] = df6["Frequencia"]
df6_medidas
```

```python
#calculo da media
media_v6 = sum(df6_medidas["Representantes"]*df6_medidas["Frequencia"])/len(df["ID"])

#calculo da variancia
variancia_v6 = sum((df6_medidas["Representantes"] - media_v6)**2)/len(df["ID"])

#calculo do desvio padrao
desvio_padrao_v6 = math.sqrt(variancia_v6)
```

```python
print("A media da variavel V6 eh {} \nA Variacia eh {} \nO Desvio Padrao eh {}".format(media_v6, variancia_v6, desvio_padrao_v6))
```
